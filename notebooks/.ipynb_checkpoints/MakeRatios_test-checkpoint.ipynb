### Now do this with the geochemical data

```python
#import the data! 

datapathlocal = "../data/tec discrim diagrams.csv"
df=pd.read_csv(datapathlocal, header=0)

data=df.dropna()
labels= data.loc[:,'Cluster_ID'] 
features = data.drop(columns=['Cluster_ID', 'Number'])
```

```python
print(features.columns)

features_wratios = GeochemFunctions.makeratios(features)

#display(features_wratios)
print('original size = ' + str(features.shape[1]))

Cnk = GeochemFunctions.numcomb(features.shape[1],2)
Rsize = features.shape[1] + Cnk

print('Checking... DID IT WORK? ? ?')
if int(Rsize) == features_wratios.shape[1]:
    print('WOOHOOO! predicted size of ' + str(int(Rsize)) + ' matches combinatorics!')
```
